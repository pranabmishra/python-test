```python
import pandas as pd
```

```python
df = pd.read_excel('./dataset/sample_pivot.xlsx')
```

```python
df.head(5)
```

```python
pivot1 = pd.pivot_table(df,index='Region')
pivot1
```

```python
pivot2 = pd.pivot_table(df,index='Region',values='Sales')
pivot2
```

```python
pivot3 = pd.pivot_table(df,index='Region',values='Sales',aggfunc = 'sum')
pivot3
```

```python
pivot4 = pd.pivot_table(df,index='Region',values='Sales',aggfunc=['sum','mean'])
pivot4
```

```python
pivot5 = pd.pivot_table(df,index='Region',values='Sales',aggfunc='sum',columns='Type',margins=True)
pivot5
```

```python
pivot5 = pd.pivot_table(df,index='Region',values='Sales',aggfunc='sum',columns='Type',margins=True,margins_name='Total')
pivot5
```
