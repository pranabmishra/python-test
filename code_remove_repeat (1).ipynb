```python
import pandas as pd
```

```python
data_set = pd.read_csv('data_mail.csv')
data_set
```

```python
# Numbers of duplicates

data_set.duplicated(subset = ['Name','Email_id']).sum()
```

```python
# Store Duplicate data_set 
# here carry all duplicates 

data_set_duplicate_1 = data_set.loc[data_set.duplicated(keep = 'first'),:] # mean 1st five rows # u can use last here
data_set_duplicate_1
```

```python
# Another method Store Duplicate data_set
# Store Duplicate data_set
# Carry perticular columns

data_set_duplicate_2 = data_set.loc[data_set.duplicated(subset = ['Name','Email_id'])]
data_set_duplicate_2
```

```python
# Save in csv file with duplicate

data_set_duplicate_2.to_csv('duplicate_data.csv')
```

```python
# Final data_set after removing duplicates

data_after_remove_repeat = data_set.drop_duplicates(keep = 'first')
data_after_remove_repeat 
```

```python
# save in csv after removing duplicate or repeat value

data_after_remove_repeat.to_csv('remove_repeat.csv')
```
