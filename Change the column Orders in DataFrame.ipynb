# How to change the column orders in the DataFrame?

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./dataset/nba.csv')
```

```python
df.head()
```

# By Using iloc and Slicer

```python
df.iloc[:,[0,1,2,8,4,5,6,7,3]]
```

# By using Loc and Column Name

```python
df.loc[:,['Name','Team','College','Position','Age','Height','Weight','Salary','Number']]
```

## Selecting few rows and 4 column

```python
df.loc[:20,['Name','Team','College','Salary']]
```

# Subsetting columns by list

```python
df[['Name', 'College', 'Team',   'Salary','Number', 'Position', 'Age', 'Height', 'Weight']]
```
