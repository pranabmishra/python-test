# Creating Subplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nobel = pd.read_csv('./dataset/nobel_prize_winner.csv')
```

```python
nobel.head(2)
```

```python
nobel_year = nobel.groupby(['Year','Category']).count()['Prize'].unstack().fillna(0).reset_index()

nobel_year
```

```python
# Creating Subplot
fig,(ax1,ax2) = plt.subplots(1,2, figsize=(18,5))

ax1.plot(nobel_year['Year'], nobel_year['Chemistry'],label='Chemistry')
ax2.plot(nobel_year['Year'], nobel_year['Medicine'], label='Medicin')


# Setting up title

ax1.set_title('Nobel Prize on Chemistry')
ax2.set_title('Nobel prize on Medicin')

ax1.legend()
ax2.legend()
plt.show()
```
